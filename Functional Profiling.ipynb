```python
import os
from gprofiler import GProfiler

def read_query_file(file_path):
    with open(file_path, 'r') as file:
        genes = file.read().splitlines()
    return genes

def run_gprofiler(genes):
    gp = GProfiler(return_dataframe=True)
    result = gp.profile(organism='hsapiens', query=genes) 
    return result

def save_results(results, output_file):
    results.to_csv(output_file, index=False)

def main(directory):
    for file_name in os.listdir(directory):
        if file_name.endswith('.txt'):
            input_file_path = os.path.join(directory, file_name)
            genes = read_query_file(input_file_path)
            
            results = run_gprofiler(genes)
            if not results.empty:
                output_file_path = os.path.join(directory, f"{file_name.replace('.txt', '')}_results.csv")
                save_results(results, output_file_path)
            else:
                print(f"No results for {file_name}")

if __name__ == "__main__":
    directory = '/Users/yazgisert/Desktop/tez/D_Cardio_Data/dominant_isoform/Maggic_Score' 
    main(directory)

```

```python
import pandas as pd
import os

def convert_csv_to_tsv(directory):
    for filename in os.listdir(directory):
        if filename.endswith('.csv'):
            csv_file_path = os.path.join(directory, filename)
            df = pd.read_csv(csv_file_path)
            
            tsv_file_path = os.path.join(directory, filename.replace('.csv', '.tsv'))
            df.to_csv(tsv_file_path, sep='\t', index=False)
            

directory = '/Users/yazgisert/Desktop/tez/D_Cardio_Data/dominant_isoform/Maggic_Score' 
convert_csv_to_tsv(directory)
```
