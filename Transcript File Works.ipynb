```python
import pandas as pd


mart_export = '/Users/yazgisert/Desktop/tez/mart_export_trans.csv'
tcga_file = '/Users/yazgisert/Desktop/tez/filtered_tcga_Kallisto_tpm.csv'
output_path = '/Users/yazgisert/Desktop/tez/new_tcga_Kallisto_tpm.csv'

mart_export = pd.read_csv(mart_export)
enst_to_gene_mapping = dict(zip(mart_export['Transcript stable ID'], mart_export['Gene stable ID']))
```

```python
tcga_data = pd.read_csv(tcga_file, dtype={0: str}, index_col=0)

def decimal_removal(enst_id):
    return str(enst_id).split('.')[0]

tcga_data.iloc[:,0] = tcga_data.iloc[:,0].apply(decimal_removal)
display(tcga_data)

tcga_data['Gene stable ID'] = tcga_data.iloc[:, 0].map(enst_to_gene_mapping)

tcga_data.to_csv(output_path, index=False)

print("Processing complete. Updated file saved as:", output_path)
```

```python
file = '/Users/yazgisert/Desktop/tez/new_tcga_Kallisto_tpm.csv'
df = pd.read_csv(file)
```

```python
df['median_expression'] = df.iloc[:, 1:-1].median(axis=1)

grouped_df = df.loc[df.groupby('Gene stable ID')['median_expression'].idxmax()]

output = '/Users/yazgisert/Desktop/tez/processed_tcga_Kallisto.csv'
grouped_df.to_csv(output, index=False)
```

```python
df = pd.read_csv('/Users/yazgisert/Desktop/tez/processed_tcga_Kallisto.csv')
martdf = pd.read_csv('/Users/yazgisert/Desktop/tez/mart_export_trans.csv')

#Mapping
gene_to_sample_mapping = df.set_index('sample')['Gene stable ID'].to_dict()
transcript_to_entrezgene = martdf.set_index('Gene stable ID')['NCBI gene (formerly Entrezgene) ID'].to_dict()

df['sample'] = df['sample'].map(gene_to_sample_mapping)
df['sample'] = df['sample'].map(transcript_to_entrezgene)
df.dropna(subset=['sample'], inplace=True)
df['sample'] = df['sample'].astype(int)


modified_df = df.drop(columns=['Gene stable ID', 'median_expression'])

modified_df.to_csv('/Users/yazgisert/Desktop/tez/last_tcga_Kallisto.csv', index=False)
```
