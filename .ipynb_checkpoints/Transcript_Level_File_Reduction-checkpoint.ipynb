```python
import pandas as pd

file1_path = '/Users/yazgisert/Desktop/tez/tcga_Kallisto_tpm.csv'  
file2 = '/Users/yazgisert/Desktop/tez/mart_export_trans.csv' 


df1 = pd.read_csv(file1_path, header=None)


df2 = pd.read_csv(file2)


ids = set(df2['Transcript stable ID'].tolist())


newdf1 = df1[df1.iloc[:, 0].isin(ids)]


newdf1.to_csv('/Users/yazgisert/Desktop/tez/filtered_tcga_Kallisto_tpm.csv', index=False, header=False)
```
